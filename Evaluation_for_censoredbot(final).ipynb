```python
# ROC of the chat filter evaluation
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
import matplotlib.pyplot as plt
import numpy as np
from sklearn.svm import SVC

classes = np.unique(chatfilter_output_test)  # Unique classes
n_classes = len(classes)

# Define and train the SVM classifier with probability estimation
svm = SVC(kernel='linear', probability=True)
svm.fit(chatfilter_input_train, chatfilter_output_train)  # Assuming `chatfilter_input_train` and `chatfilter_output_train` are defined

# Binarize the labels for one-vs-all computation
y_test_binarized = label_binarize(chatfilter_output_test, classes=classes)
y_score = svm.predict_proba(chatfilter_input_test)  # Assuming `svm` is your classifier

# Plot the ROC curve for each class
plt.figure()
for i in range(n_classes):
    fpr, tpr, _ = roc_curve(y_test_binarized[:, i], y_score[:, i])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, lw=2, label=f'Class {classes[i]} (area = {roc_auc:.2f})')

# Plot settings
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC of chat filter')
plt.legend(loc="lower right")
plt.show()
```
